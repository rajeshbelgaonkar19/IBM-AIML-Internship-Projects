# Polynomial Regression

## Importing the libraries

```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Importing the dataset

```
dataset = pd.read_excel('Energy.xlsx')

x = dataset.iloc[:, [2]].values
y = dataset.iloc[:, 3].values
print("X values (3rd column):\n", x)
print("Y values (4th column):\n", y)

```

## Training the Linear Regression model on the whole dataset

```
from sklearn.linear_model import LinearRegression
lin_reg=LinearRegression()
lin_reg.fit(x, y)
```

## Training the Polynomial Regression model on the whole dataset

```
from sklearn.preprocessing import PolynomialFeatures
pf=PolynomialFeatures(degree=5)
x_poly=pf.fit_transform(x)
poly_reg= LinearRegression()
poly_reg.fit(x_poly,y)
```

```
print(x_poly)
```

## Visualising the Linear Regression results

```
plt.scatter(x, y, color='blue')
plt.plot(x,lin_reg.predict(x), color='red')
plt.xlabel("Low-carbon electricity")
plt.ylabel("GDP Growth")
plt.title("Low-carbon electricity vs GDP Growth")
plt.show()
```

## Visualising the Polynomial Regression results

```
plt.scatter(x, y, color='blue')
plt.plot(x,poly_reg.predict(x_poly), color='red')
plt.xlabel("Low-carbon electricity")
plt.ylabel("GDP Growth")
plt.title("Low-carbon electricity vs GDP Growth")
plt.show()
```

## Predicting a new result with Linear Regression

```
lin_reg.predict([[6.5]])
```

## Predicting a new result with Polynomial Regression

```
poly_reg.predict(pf.fit_transform([[6.5]]))
```
